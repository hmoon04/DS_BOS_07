```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from numpy import nan as NA
```

```python
dF = pd.read_csv("../Data/libor.csv")
dF.head()
```

#### There are six variables, all indexed by time - i.e. all are time series. 
#### Look to merge the time variables into a single common time index

```python
dFC = dF.dropna()
```

```python
for name in dFC.columns:
    print name
```

```python
dFC.head()
```

### One way to achieve this is to take each time series separately as a pair of vectors

```python
lib1=dFC['LIBOR_date']
lib2=dFC['LIBOR_price'].values
```

```python
oil1 = dFC['Oil_date']
oil2 = dFC['Oil_price'].values
```

```python
ois1 = dFC['OIS_date']
ois2 = dFC['OIS_price'].values
```

```python
usei1 = dFC['US_equity_index_date']
usei2 = dFC['US_equity_index'].values
```

```python
us10Y1 = dFC['US10Y_date']
us10Y2 = dFC['US10Y_yield'].values
```

```python
usTW1 = dFC['USD_trade_weighted_date']
usTW2 = dFC['USD_trade_weighted_index'].values
```

#### Using the pandas data conversion function intrepret the date strings.
#### Then create 6 separate time series, all formally indexed by date.

```python
lib_s = [pd.datetime.strptime(x, '%m/%d/%Y') for x in lib1]
oil_s = [pd.datetime.strptime(x, '%m/%d/%Y') for x in oil1]
ois_s = [pd.datetime.strptime(x, '%m/%d/%Y') for x in ois1]
usei_s =[pd.datetime.strptime(x, '%m/%d/%Y') for x in usei1]
us10Y_s =[pd.datetime.strptime(x, '%m/%d/%Y') for x in us10Y1]
usTW_s =[pd.datetime.strptime(x, '%m/%d/%Y') for x in usTW1]

libS = pd.Series(lib2, index=lib_s)
oisS = pd.Series(ois2, index=ois_s)
oilS = pd.Series(oil2, index=oil_s)
useiS = pd.Series(usei2, index=usei_s)
us10YS = pd.Series(us10Y2, index=us10Y_s)
usTWS = pd.Series(usTW2, index=usTW_s)
```

#### Using the date indices recombine the variables into a single data frame.
#### All variables are now date aligned.

```python
df_Final=pd.concat([libS,oisS,oilS,useiS,us10YS,usTWS], axis=1)
df_Final.head()
```

### NA's have arisen because of the time alignment - some colummns did not have entries for times that other columns did.
### Note also that the time series has no period, some days are missing

```python
print df_Final.index[0]
print df_Final.index[-1]
print df_Final.index[-1] - df_Final.index[0]
print len(df_Final.index)
```

### Resample for business days

```python
df_Final2 = df_Final.resample('B')
print df_Final2.index[0]
print df_Final2.index[-1]
print df_Final2.index[-1] - df_Final2.index[0]
print len(df_Final2.index)
```

```python
df_Final2.info()
```

##### So there was only 1 day missing!!

```python
df_Final3 = df_Final2.ffill()
```

```python
df_Final3.info()
```

```python
print df_Final3.index[0]
print df_Final3.index[-1]
print df_Final3.index[-1] - df_Final3.index[0]
print len(df_Final3.index)
```

### Visualize the time aligned data

```python
df_Final3.columns = ['LIB_Price', 'OIS_Price', 'Oil_Price', 'EIndex', 'Yield', 'TWI']
df_Final3.plot(subplots=True, figsize=(20,20))
plt.show()
```
